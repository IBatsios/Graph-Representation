```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.Graph()
```

```python
H = nx.Graph()
```

```python
H.add_nodes_from([3], degree_centrality='0.333')
```

```python
H.add_nodes_from([4], degree_centrality='0.445')
```

```python
H.add_nodes_from([2], degree_centrality='0.233')
```

```python
H.add_edges_from([(1,2), (2,3), (3,4), (1,4)])
```

```python
nx.draw_networkx(H,node_size=800,with_labels=True, node_color=range(4), cmap=plt.cm.Oranges)
```

```python
dc = nx.degree_centrality(H)
```

```python
print(dc.items())
```

```python
nx.set_node_attributes(H, dc.items())
```
